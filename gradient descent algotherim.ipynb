```python
import matplotlib.pyplot as plt #LoTex Markdown and generating data with Numpy
import numpy as np
%matplotlib inline
```

$$f(x)=x^2+x+1$$  

## squre of hyperbola eqn
(cell type in mark-cell)

```python
def f(x):
    return x**2+x+1
```

```python
x_1=np.linspace(start=-3,stop =3,num=100) #generating data using np.linespace
```

```python
type(x_1)
```

```python
print(x_1)
```

```python
plt.plot(x_1,f(x_1))
plt.show()
```

```python
plt.xlim([-3,3])
plt.ylim([0,8])
plt.title('cost function',fontsize=20)
plt.xlabel('x',fontsize=20)
plt.ylabel('f(x)',fontsize=20)
plt.plot(x_1,f(x_1))
plt.show()
```

```python
def df(x):   #plotting the derivtie 
    return 2*x + 1 #derivative of f(x)=x^2 + x+1
```

```python
plt.figure(figsize=[15,5]) #plot function and derivate side by side
#cost function
plt.subplot(1,2,1) #plot subplot(row,col,index)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.title('cost function ',fontsize=20)
plt.xlabel('x',fontsize=20)
plt.ylabel('f(x)',fontsize=20)
plt.grid()
plt.plot(x_1,f(x_1),color='r',linewidth=3)
#derivative of cost fonction
plt.subplot(1,2,2)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.title('slope of the cost function' , fontsize=17)

plt.xlabel('x',fontsize=20)
plt.ylabel('df(x)',fontsize=16)
plt.grid()

plt.plot(x_1,df(x_1),color='g',linewidth=2)
plt.show()
```

```python
new_location=3
previous_loc=0
step_size=0.1

for i in range(25):
    previous_loc=new_location
    slope=df(previous_loc)
    new_location=previous_loc-step_size * slope
print(" local minimum occures at", new_location)
print("slope at this point is",df(new_location))
print("cost at this point is",f(new_location))
```

```python
new_x=3
prev_x=0
step_size=0.1

for i in range(5):
    prev_x=new_x
    slope=df(prev_x)
    new_x=prev_x - step_size * slope
    drop = abs(new_x - prev_x)          #?
    print(i,drop,'\n')
print(" local minimum occures at", new_x)
print("slope at this point is",df(new_x))
print("cost at this point is",f(new_x))
```

```python
# introdusing precision to terminate the loop
new_x=3
prev_x=0
step_size=0.1
precision=0.0001  #to reduse itteration

for i in range(500):
    prev_x=new_x
    slope=df(prev_x)
    new_x=prev_x - step_size * slope
    drop = abs(new_x - prev_x)          
    if drop < precision:
        print('the loop ran ' + str(i)+ 'times','\n')  #condition for termaniting the loop / n to string for printing / terminatig loop
        break
        
print(" local minimum occures at", new_x)
print("slope at this point is",df(new_x))
print("cost at this point is",f(new_x))         # why 40?
```

```python
#visualising , storing the x 's  and the slope1 at x's in a list
new_x=3
prev_x=0
step_size=0.1
precision=0.00001  
x_list=[new_x]  #?
slope_list=[df(new_x)]  #?

for i in range(500):
    prev_x=new_x
    slope1 =df(prev_x)
    new_x=prev_x - step_size * slope1
    drop = abs(new_x - prev_x)   
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if drop < precision:
        print('the loop ran ' + str(i)+ 'times','\n')
        break
        
print(" local minimum occures at", new_x)
print("slope1 at this point is",df(new_x))
print("cost at this point is",f(new_x))  
```
